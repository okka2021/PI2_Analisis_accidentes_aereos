```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_aereo = pd.read_csv('AccidentesAvionesetl.csv')
```

```python
df_aereo.head()
```

vamos a ver las realciones entre variables

```python
sns.pairplot(data=df_aereo);
```

grafica para ver la cantidad de accidentes por año 

```python
acc_años = df_aereo['anio'].value_counts().sort_index()

plt.figure(figsize=(10, 4))
sns.lineplot(x=acc_años.index, y=acc_años.values, marker='o', markersize=8, color='red', linewidth=1)
plt.title('ACCIDENTES OCURRIDOS POR AÑO', fontsize=16)
plt.xlabel('AÑO', fontsize=16)
plt.ylabel('Cantidad de accidentes', fontsize=8)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=1)
plt.tight_layout()

plt.show()
```

grafica para ver la cantidad de el total de personas a bordo con los fallecidos en tierra 

```python
plt.figure(figsize=(16, 12))
plt.scatter(df_aereo['total_abordo'], df_aereo['fallecidos_en_tierra'])
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de dispersión')
plt.show()
```

grafica para ver la catidad de accidentes por ubicacion

```python
acc_ubi = df_aereo['ubicacion'].value_counts().head(30)
plt.figure(figsize=(12, 6))
acc_ubi.plot(kind='bar', color='green')
plt.xlabel('UBICACION')
plt.ylabel('CANTIDAD DE ACCIDENTES')
plt.title('ACCIDENTES POR UBICACION')
plt.show()
```

```python
grafica para ver la catidad de accidentes por aerolinea
```

```python
acc_aero = df_aereo['aerolinea'].value_counts().head(30)
plt.figure(figsize=(12, 6))
acc_aero.plot(kind='bar', color='red')
plt.xlabel('AEROLINEA')
plt.ylabel('CANTIDAD DE ACCIDENTES')
plt.title('ACCIDENTES POR AEROLINEA')
plt.show()
```

grafica para ver la catidad de accidentes por fecha

```python
acc_fec = df_aereo['fecha'].value_counts().head(30)
plt.figure(figsize=(12, 6))
acc_fec.plot(kind='bar', color='yellow')
plt.xlabel('FECHA DE SINIESTRO')
plt.ylabel('CANTIDAD DE ACCIDENTES')
plt.title('ACCIDENTES POR FECHA')
plt.show()
```

importacion de librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

lectura de archivo suministrado para ver la data

```python
df_aereo = pd.read_csv('AccidentesAviones.csv')
df_aereo.info()
```

normalizacion de nombres de columnas

```python
df_aereo.rename(columns={'Unnamed: 0':'id'}, inplace=True)
df_aereo.rename(columns={'HORA declarada':'hora_siniestro'}, inplace=True)
df_aereo.rename(columns={'Ruta':'ubicacion'}, inplace=True)
df_aereo.rename(columns={'OperadOR':'aerolinea'}, inplace=True)
df_aereo.rename(columns={'flight_no':'id_vuelo'}, inplace=True)
df_aereo.rename(columns={'route':'ruta'}, inplace=True)
df_aereo.rename(columns={'ac_type':'tipo'}, inplace=True)
df_aereo.rename(columns={'registration':'modelo'}, inplace=True)
df_aereo.rename(columns={'all_aboard':'total_abordo'}, inplace=True)
df_aereo.rename(columns={'PASAJEROS A BORDO':'pasajeros'}, inplace=True)
df_aereo.rename(columns={'crew_aboard':'tripulación'}, inplace=True)
df_aereo.rename(columns={'cantidad de fallecidos':'fallecidos'}, inplace=True)
df_aereo.rename(columns={'passenger_fatalities':'fallecidos_pasajeros'}, inplace=True)
df_aereo.rename(columns={'crew_fatalities':'fallecidos_tripulacion'}, inplace=True)
df_aereo.rename(columns={'ground':'fallecidos_en_tierra'}, inplace=True)
df_aereo.rename(columns={'summary':'resumen'}, inplace=True)
df_aereo.head(3)
```

eliminacion de columnas que no son relevantes para el analisis

```python
df_aereo.drop(['id_vuelo', 'modelo', 'cn_ln', 'resumen', 'hora_siniestro','pasajeros',	'tripulación',	'fallecidos_pasajeros',	'fallecidos_tripulacion'], axis=1, inplace=True)
```

```python
df_aereo.head
```

```python
df_aereo.head()
```

muchos datos estan con el signo '?' se cambian por valores NaN y asi poder saber cuando hay 

```python
df_aereo.replace('?', np.nan, inplace=True)
```

```python
df_aereo.head()
```

se revisa cuantos nan, en que columnas hay y las filtro para un mejor analisis

```python
df_aereo.isna().sum()
```

```python
df1=df_aereo[df_aereo.isnull().any(axis=1)]
df1.head(10)
```

En el analisis se ve que en las columnas ubicacion, aerolinea, ruta, tipo, hay valores NaN que seran cambiados por Sin informacion,

```python
df_aereo['ubicacion'] = df_aereo['ubicacion'].replace(np.nan, 'sin informacion')
df_aereo['aerolinea'] = df_aereo['aerolinea'].replace(np.nan, 'sin informacion')
df_aereo['ruta'] = df_aereo['ruta'].replace(np.nan, 'sin informacion')
df_aereo['tipo'] = df_aereo['tipo'].replace(np.nan, 'sin informacion')
```

```python
en las columnas total_abordo, fallecidos y  fallecidos en tierra elimino las filas si y solo si las tres contienen datos NaN
```

```python
df_aereo=df_aereo.dropna(subset= ['total_abordo','fallecidos','fallecidos_en_tierra'], how='all')
```

```python
df_aereo.isna().sum()
```

cambio los datos NaN por 0 en la columnas total a bordo y fallecidos e tierra

```python
df_aereo= df_aereo.fillna (0)
```

```python
df_aereo.isna().sum()
```

```python
df_aereo.head(20)
```

```python
df_aereo.info()
```

cabiar los tipo de fecha a datatime, total_abordo,fallecidos y fallecidos en tierra a int

```python
df_aereo['fecha'] = pd.to_datetime(df_aereo['fecha'])
df_aereo['fallecidos'] = df_aereo['fallecidos'].astype(int)
df_aereo['fallecidos_en_tierra'] = df_aereo['fallecidos_en_tierra'].astype(int)
df_aereo['total_abordo'] = df_aereo['total_abordo'].astype(int)
```

```python
df_aereo.info()
```

creo una columna de años 

```python
df_aereo['anio'] = pd.to_datetime(df_aereo['fecha']).dt.year
```
